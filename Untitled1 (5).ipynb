```python
# imports S&P 500 data from 2000-2024

df= pd.read_csv('bro.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'clipped' DataFrame is already defined and contains the historical close prices

# Step 1: Calculate historical log returns
clipped['Returns'] = np.log(1 + clipped['Close'].pct_change())
returns = clipped['Returns'].dropna()

# Step 2: Estimate mean and standard deviation of returns
mu, sigma = returns.mean(), returns.std()

# Step 3: Simulate future price paths
num_simulations = 10000
num_days = len(returns)
initial_price = clipped['Close'].iloc[-1]

simulated_drawdowns = []

for _ in range(num_simulations):
    # Simulate daily returns
    simulated_returns = np.random.normal(mu, sigma, num_days)
    
    # Simulate price path
    simulated_prices = initial_price * np.exp(np.cumsum(simulated_returns))
    
    # Calculate cumulative returns
    cumulative_returns = (1 + simulated_returns).cumprod()
    
    # Calculate rolling maximum
    rolling_max = np.maximum.accumulate(cumulative_returns)
    
    # Calculate drawdown
    drawdowns = (cumulative_returns / rolling_max) - 1
    
    # Find maximum drawdown
    max_drawdown = drawdowns.min()
    simulated_drawdowns.append(max_drawdown)

# Step 4: Define threshold for maximum drawdown
historical_cumulative_returns = (1 + returns).cumprod()
historical_rolling_max = np.maximum.accumulate(historical_cumulative_returns)
historical_drawdowns = (historical_cumulative_returns / historical_rolling_max) - 1
threshold_drawdown = historical_drawdowns.min()

# Step 5: Calculate the probability of exceeding the threshold
exceed_threshold = np.sum(np.array(simulated_drawdowns) < threshold_drawdown) / num_simulations

# Results
print(f"Threshold Drawdown: {threshold_drawdown:.2%}")
print(f"Probability of Exceeding Threshold Drawdown: {exceed_threshold:.2%}")


```
